```python
import duckdb
duckdb.sql("""
    SELECT customer_sk, SUM(amount) AS total_sales
    FROM '../data/fact/fact_sales.csv'
    GROUP BY customer_sk
    ORDER BY total_sales DESC
    LIMIT 5
""").df()
```

```python
duckdb.sql("""
    SELECT 
        c.name AS customer_name, 
        SUM(f.amount) AS total_sales
    FROM 
        '../data/fact/fact_sales.csv' f
    JOIN 
        '../data/dim/dim_customer.csv' c 
    ON 
        f.customer_sk = c.customer_sk
    GROUP BY 
        c.name
    ORDER BY 
        total_sales DESC
    LIMIT 10
""").df()
```
